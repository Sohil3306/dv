```python
import nltk
from nltk.corpus import movie_reviews
from nltk.classify import NaiveBayesClassifier, accuracy
from nltk import bigrams

nltk.download("movie_reviews")

def extract_features(words):
    feats = {w: True for w in words}
    feats.update({b: True for b in bigrams(words)})
    return feats

data = [(extract_features(movie_reviews.words(f)), c)
        for c in movie_reviews.categories()
        for f in movie_reviews.fileids(c)]

train_data, test_data = data[:1600], data[1600:]
clf = NaiveBayesClassifier.train(train_data)

print(f"Accuracy: {accuracy(clf, test_data) * 100:.2f}%")
pred = clf.classify(extract_features(input("Enter a sentence: ").split()))
print(f"The sentence is classified as: {pred}")
```
