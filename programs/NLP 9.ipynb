```python
import nltk
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")

text = input("Enter a sentence to process: ")
tagged = nltk.pos_tag(nltk.word_tokenize(text))

# Chunk grammar
chunker = nltk.RegexpParser(r"""
    NP: {<DT|JJ>*<NN>}   # Chunk
""")
chunked = chunker.parse(tagged)
print("\nChunked Sentence:\n", chunked)

# Chink grammar (fixed)
chinker = nltk.RegexpParser(r"""
    NP: {<DT|JJ>*<NN>}   # Chunk
    NP: }<IN>{           # Chink prepositions
""")
chinked = chinker.parse(tagged)
print("\nChinked Sentence:\n", chinked)

chunked.draw()
```
