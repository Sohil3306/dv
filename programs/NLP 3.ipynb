```python
import nltk
from nltk.corpus import wordnet
# Download required NLTK data
nltk.download('wordnet')
nltk.download('omw-1.4')
# Function to get synonyms and antonyms
def get_synonyms_antonyms(word):
    synonyms = set()
    antonyms = set()
    # Retrieve all synsets (senses) of the word
    for syn in wordnet.synsets(word):
        # Add all lemmas of the word to synonyms
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
            # Check if the lemma has antonyms and add them
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())
    return synonyms, antonyms
# Input word from the user
word = input("Enter a word: ")
# Get synonyms and antonyms
synonyms, antonyms = get_synonyms_antonyms(word)
# Output results
print(f"Synonyms of '{word}': {', '.join(synonyms) if synonyms else 'None'}")
print(f"Antonyms of '{word}': {', '.join(antonyms) if antonyms else 'None'}")
```
