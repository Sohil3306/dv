```python
import matplotlib.pyplot as plt
import numpy as np
# Sample monthly sales data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
"Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [120, 150, 180, 220, 200, 250, 300, 270, 260, 230, 210, 190]
# Find peak sales
max_idx = np.argmax(sales)
max_month = months[max_idx]
max_value = sales[max_idx]
# Create subplot layout
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Telling a Story and Making a Point with Data", fontsize=16, fontweight="bold")
# 1. Just showing data (plain chart)
axes[0,0].plot(months, sales, marker="o", color="gray")
axes[0,0].set_title("1. Raw Data (No Story)")
axes[0,0].set_ylabel("Sales")
# 2. Highlighting key point
axes[0,1].plot(months, sales, marker="o", color="skyblue", linewidth=2)
axes[0,1].scatter(max_month, max_value, color="red", s=100, zorder=5, label="Peak Sales")
axes[0,1].set_title("2. Highlight Key Insight")
axes[0,1].legend()
# 3. Adding a trendline to show growth/decline
x = np.arange(len(months))
coeffs = np.polyfit(x, sales, 1)
trend = np.poly1d(coeffs)
axes[1,0].plot(months, sales, marker="o", label="Sales")
axes[1,0].plot(months, trend(x), "r--", label="Trendline")
axes[1,0].set_title("3. Show Trend with Story")
axes[1,0].legend()
# 4. Annotated storytelling
axes[1,1].plot(months, sales, marker="o", color="green", linewidth=2)
axes[1,1].scatter(max_month, max_value, color="red", s=100)
axes[1,1].annotate(f"Peak: {max_value} in {max_month}",
xy=(max_idx, max_value),
xytext=(max_idx-3, max_value+30),
arrowprops=dict(facecolor='red', shrink=0.05),
fontsize=11, color="red")
axes[1,1].set_title("4. Annotated Storytelling")
# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
