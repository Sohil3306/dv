```python
from nltk.util import ngrams 
def unigram (sentence): 
    n = 1 
    unigram=ngrams(sentence.split(),n) 
    for i in unigram: 
        print(i) 
def bigram (sentence): 
     n = 2 
     bigram= ngrams(sentence.split(),n) 
     for  i in bigram: 
         print(i) 
def trigram (sentence): 
    n = 3 
    trigram = ngrams(sentence.split(),n) 
    for i in trigram: 
         print(i) 
def ngram(sentence,n): 
    ngram_sentence=ngrams(sentence.split(),n) 
    for i in ngram_sentence: 
        print(i) 
sentence = input("enter sentence:") 
choice=True 
while choice: 
    option=input("Enter option A-Unigram,B-Bigram,C-trigram,D-Ngram,E- Exit").upper() 
    if option=="A": 
        unigram(sentence) 
    elif option=="B": 
        bigram(sentence) 
    elif option=="C": 
        trigram(sentence) 
    elif option=="D": 
        n=int(input("Enter n:")) 
        ngram(sentence,n) 
    elif option=="E": 
        choice=False
```
