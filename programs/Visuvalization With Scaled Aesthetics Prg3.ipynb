```python
# Progarm No:3
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Step 1: Create sample employee dataset
data = pd.DataFrame({
    'EmployeeID': range(1, 26),
    'Experience': [1, 3, 5, 7, 2, 4, 6, 8, 10, 12, 1, 4, 6, 9, 11, 2, 3, 5, 7, 8, 10, 12, 4, 6, 9],
    'Age': [22, 25, 28, 32, 24, 27, 30, 35, 40, 45, 23, 26, 29, 33, 38, 24, 26, 31, 34, 37, 41, 46, 25, 28, 36],
    'Salary': [30000, 40000, 50000, 60000, 35000, 45000, 55000, 65000, 75000, 85000,
               32000, 42000, 52000, 62000, 72000, 34000, 36000, 56000, 66000, 76000, 80000, 90000, 43000, 53000, 70000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'Finance', 'IT', 'HR',
                   'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'Finance', 'IT',
                   'HR', 'Finance', 'IT', 'HR', 'Finance']
})

# Step 2: Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(22, 7))

# Scatter Plot 1: Experience vs Age
sns.scatterplot(
    x='Experience', y='Age', size='Salary', hue='Department',
    data=data, sizes=(100, 1200), alpha=0.7, palette='Set2', ax=axes[0]
)
axes[0].set_title('Experience vs Age', fontsize=14)
axes[0].grid(True)

# Scatter Plot 2: Experience vs Salary
sns.scatterplot(
    x='Experience', y='Salary', size='Age', hue='Department',
    data=data, sizes=(100, 1200), alpha=0.7, palette='Set1', ax=axes[1]
)
axes[1].set_title('Experience vs Salary', fontsize=14)
axes[1].grid(True)

# Scatter Plot 3: Age vs Salary
sns.scatterplot(
    x='Age', y='Salary', size='Experience', hue='Department',
    data=data, sizes=(100, 1200), alpha=0.7, palette='Set3', ax=axes[2]
)
axes[2].set_title('Age vs Salary', fontsize=14)
axes[2].grid(True)

# Step 3: Final plot adjustments
plt.tight_layout()
plt.suptitle('Multiple Scatter Plots with Scaled Aesthetics (Employee Data)', fontsize=18, y=1.05)
plt.show()
```
