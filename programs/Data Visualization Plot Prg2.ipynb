```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
revenue = [10000, 15000, 18000, 13000, 20000, 22000]
sales = [120, 135, 160, 145, 190, 210]
product_prices = np.random.normal(50, 10, 100)

categories = ['Electronics', 'Furniture', 'Clothing', 'Grocery']
category_sales = [40000, 25000, 30000, 15000]

regions = ['North', 'South', 'East', 'West']
market_share = [30, 25, 20, 25]

ratings = np.random.normal(4, 0.5, 100)

ad_spend = np.random.randint(1000, 10000, 50)
monthly_revenue = ad_spend * 2 + np.random.randint(1000, 3000, 50)

# 1. Line Chart - Monthly Revenue
plt.figure(figsize=(6, 4))
plt.plot(months, revenue, marker='o', color='blue')
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue ($)') 
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Area Chart - Cumulative Sales
plt.figure(figsize=(6, 4))
plt.fill_between(months, np.cumsum(sales), color='skyblue', alpha=0.5)
plt.plot(months, np.cumsum(sales), color='blue', label='Cumulative Sales')
plt.title('Cumulative Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 3. Histogram - Product Price Distribution
plt.figure(figsize=(6, 4))
plt.hist(product_prices, bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Product Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 4. Bar Chart - Sales by Product Category
plt.figure(figsize=(6, 4))
plt.bar(categories, category_sales, color='green')
plt.title('Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.tight_layout()
plt.show()
```

```python
# 5. Pie Chart - Market Share by Region
plt.figure(figsize=(6, 4))
plt.pie(market_share, labels=regions, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Market Share by Region')
plt.tight_layout()
plt.show()
```

```python
# 6. Box Plot - Customer Ratings
plt.figure(figsize=(6, 4))
plt.boxplot(ratings, vert=False)
plt.title('Customer Ratings Distribution')
plt.xlabel('Rating')
plt.tight_layout()
plt.show()
```

```python
# 7. Scatter Plot - Ad Spend vs Revenue
plt.figure(figsize=(6, 4))
plt.scatter(ad_spend, monthly_revenue, color='purple', alpha=0.7)
plt.title('Ad Spend vs Monthly Revenue')
plt.xlabel('Advertising Spend ($)')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
