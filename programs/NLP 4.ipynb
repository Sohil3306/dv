```python
# Install and download necessary NLTK data
import nltk
from nltk.corpus import stopwords
from collections import Counter
import re
# Download the stopwords data from NLTK
nltk.download('stopwords')
# Function to get the most frequent words, excluding stopwords
def get_most_frequent_words(text, n=50):
    # Convert text to lowercase and remove non-alphabetic characters
    words = re.findall(r'\b\w+\b', text.lower())
    # Filter out stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]
    # Count word frequencies
    word_counts = Counter(filtered_words)
    # Get the 'n' most common words
    most_common_words = word_counts.most_common(n)
    return most_common_words
# Input text corpus
text = input("Enter the text corpus: ")
# Get the 50 most frequent non-stopwords
top_words = get_most_frequent_words(text, 50)
# Display the results
print("50 Most Frequently Occurring Words (Excluding Stopwords):")
for word, freq in top_words:
    print(f"{word}: {freq}")
```
