```python
import nltk 
from nltk.tag import UnigramTagger 
from nltk.tag.brill import Pos, Template 
from nltk.tag.brill_trainer import BrillTaggerTrainer 
from nltk.corpus import treebank 
nltk.download("treebank") 
data = treebank.tagged_sents() 
train_data = data[:3000] 
test_data = data[3000:] 
unigram_tagger = UnigramTagger(train_data) 
templates = [ 
Template(Pos([-1])),   # Previous word 
Template(Pos([1])),    # Next word 
Template(Pos([-2])),   # Two words before 
Template(Pos([2])),    # Two words after 
Template(Pos([-1, 1])) # Surrounding words 
] 
trainer = BrillTaggerTrainer(initial_tagger=unigram_tagger, templates=templates) 
brill_tagger = trainer.train(train_data, max_rules=10)  # Limit to 10 rules for simplicity 
accuracy = brill_tagger.evaluate(test_data) 
print(f"Brill Tagger Accuracy: {accuracy * 100:.2f}%") 
input_sentence = input("Enter a sentence to tag: ").split() 
tagged_sentence = brill_tagger.tag(input_sentence) 
print("\nTagged Sentence:") 
print(tagged_sentence)
```
