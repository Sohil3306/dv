```python
import nltk
from nltk import PorterStemmer, LancasterStemmer
from nltk.stem import WordNetLemmatizer
import string

# Download required NLTK resources
nltk.download('punkt')        
nltk.download('punkt_tab')    
nltk.download('wordnet')      
nltk.download('omw-1.4')      
# Function to normalize text
def normalize_text(text):
    # Step 1: Convert the text to lowercase
    text = text.lower()
    # Step 2: Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text
# User input
raw_text = input("Enter some text to normalize: ")
# Tokenize input text
tokens = nltk.word_tokenize(raw_text)
# Process text
normalized_text = normalize_text(raw_text)
# Create stemmer and lemmatizer objects
porter = PorterStemmer()
lancaster = LancasterStemmer()
wnl = WordNetLemmatizer()
# Apply stemming and lemmatization
porter_stems = [porter.stem(t) for t in tokens]
lancaster_stems = [lancaster.stem(t) for t in tokens]
lemmatized_tokens = [wnl.lemmatize(t) for t in tokens]
# Output results
print("\n--- RESULTS ---")
print("Normalized Text:", normalized_text)
print("Word Tokenization:", tokens)
print("Porter Stemmer:", " ".join(porter_stems))
print("Lancaster Stemmer:", " ".join(lancaster_stems))
print("Lemmatization:", " ".join(lemmatized_tokens))
```
